```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
# saving
saving_rate = 0.60
#income
income=50000
#yearly expensive
expenses = income * (1-saving_rate)
#interest on assets
interest_rate = 0.08
# amount to invest
investment = income-expenses
# expectied annual return
annual_return = investment * interest_rate
# current year
year = 2024
# get withdrawal rate
withdrawal_rate = 0.04
```

```python
# create empty list data set
Year =[]
Yearly_income=[]
Yearly_expenses=[]
Yearly_invested_amount=[]
Total_invested_amount=[]
Annual_returns=[]
```

```python
#Append data to list
Year.append(year)
Yearly_income.append(income)
Yearly_expenses.append(expenses)
Yearly_invested_amount.append(income-expenses)
Total_invested_amount.append(investment)
Annual_returns.append(annual_return)
```

```python
# variable to invested years/row
invested_years=30
for i in range(0,invested_years -1):
    investment = (income-expenses)+annual_return + investment
    annual_return = investment * interest_rate
    # append the data to our list
    Year.append(year + i+1)
    Yearly_income.append(income)
    Yearly_expenses.append(expenses)
    Yearly_invested_amount.append(income-expenses)
    Total_invested_amount.append(investment)
    Annual_returns.append(annual_return)
```

```python
# create datset
df=pd.DataFrame()
df['Year']=Year
df['Yearly_income']=Yearly_income
df['Yearly_expenses']=Yearly_expenses
df['Yearly_invested_amount']=Yearly_invested_amount
df['Total_invested_amount']=Total_invested_amount
df['Annual_returns']=Annual_returns
Yearly_withdrawal_amount=np.array(Total_invested_amount)*withdrawal_rate
df['Yearly_withdrawal_amount']=Yearly_withdrawal_amount
```

```python
df
```

```python
#plot
plt.plot(df['Year'], df['Yearly_expenses'],label = 'Yearly_expenses')
plt.plot(df['Year'], df['Yearly_withdrawal_amount'],label = 'Yearly_withdrawal_amount')
plt.xlabel('Years')
plt.ylabel('USD ($)')
plt.xticks(df['Year'],rotation = 45)
plt.title('Yaers until Retirement')
plt.legend()
plt.show()
```
